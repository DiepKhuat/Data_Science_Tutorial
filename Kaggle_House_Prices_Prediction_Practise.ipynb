## 2.Feature Selections:

```python
# Pick sẵn rồi
# Make a list of features that impact on salesprice
features = ['LotArea', 'YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd'] 
```

## 3. Splitting dataset into X and y

```python
X =data[features]
y=data["SalePrice"]
```

```python
data.head()
```

```python
X.head()
```

```python
y.head()
```

#### X,y → X_train,y_train, X_valid,y_valid

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y, train_size=0.8, test_size=0.2,random_state=0) # fix seed 0 cho random
```

### 4. Training Machine Learning Model

```python
from sklearn.tree import DecisionTreeRegressor
# gọi cái class này ra, initiate khởi động cái model trên
# DecisionTreeRegressor() gọi như này ok but pass random seed vào thì consistant
dt_model= DecisionTreeRegressor(random_state=1)
```

```python
# Đưa data vào Fit training data into model
dt_model.fit(X_train, y_train)
```

```python
# Kết quả trên là đã fit train xong, bh sẽ predict với X_valid 
y_predict = dt_model.predict(X_valid.head()) # vì X_valid có 292 nên print frist 5 rows
```

```python
y_predict
```

```python
# Check y_valid (gtri thật của 20%) với y_predict( dc xây dựng từ X_valid(20%))
pd.DataFrame({'y':y_valid.head(), 'y_preds':y_predict})
```

```python
# Làm thêm bằng module mới
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(X_train, y_train)
```

```python
rf_model_preds = dt_model.predict(X_valid)
```

### 5. Model Evaluation

```python
# Calculate the mean absolute error MAE of your Random Forest model on the validation data
from sklearn.metrics import mean_absolute_error
rf_model_mae = mean_absolute_error(rf_model_preds,y_valid)
```

```python
rf_model_on_full_data = RandomForestRegressor()
rf_model_on_full_data.fit(X,y)
```
